```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/healthcare_dataset.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df['Date of Admission'] = pd.to_datetime(df['Date of Admission'])
df['Discharge Date'] = pd.to_datetime(df['Discharge Date'], dayfirst=True)
```

```python
plt.hist(df['Gender'],log = True )
plt.xlabel('Gender')
plt.ylabel('Frequency')
plt.title('Gender Distribution')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Age'], bins=50, kde=True, color="green",  line_kws={'color': 'dark', 'lw': 6, 'ls': '--'})
plt.title('Distribution of age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

```

```python
blood_type_counts = df['Blood Type'].value_counts()

plt.figure(figsize=(10,6))

for i, value in enumerate(blood_type_counts.values):
    plt.text(i, value + 0.1, str(value), ha='center')

plt.bar(blood_type_counts.index, blood_type_counts.values, width=0.5, color='red')
plt.xlabel('Blood Type')
plt.ylabel('Frequency')
plt.title('Blood Type Distribution')
plt.show()
```

```python
medical_condition_counts = df['Medical Condition'].value_counts()

plt.figure(figsize=(10,6))

plt.pie(medical_condition_counts.values, labels=medical_condition_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Medical Condition Distribution')
plt.show()
```

```python
pd.crosstab(df['Medical Condition'], df['Test Results'])
sns.heatmap(pd.crosstab(df['Medical Condition'], df['Test Results']), annot=True, cmap='Blues', fmt='g')
plt.title('Medical Condition vs Test Results')
plt.show()
```


the results of the heatmap shows the number chart of different medical conditions and their abnormality/normality/inconclusiveness
the box with more intensity shows greater number of cases belonging to that particular medical condition
